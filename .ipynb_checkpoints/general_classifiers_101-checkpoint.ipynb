#### Helper Functions and Common Parameters 

```python
def calc_model_acc(y_true,y_pred,sample): 
    """returns model accuracy results calculated based on true versus predicted labels"""
    from sklearn.metrics import confusion_matrix
    from sklearn.metrics import accuracy_score
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    tpr = round(tp/(tp+fn),4); tnr = round(tn/(tn+fp), 4)
    acc = round(accuracy_score(y_true, y_pred)*100,2)
    _r_format = '{:<5d}|{:<5d}|{:<5d}|{:<5d}|{:<7.2f}%|{:<5.2f}|{:<5.2f}'
    return _r_format.format(tp,fp,tn,fn,acc,tpr,tnr)

_format = '{:5s}|{:5s}|{:5s}|{:5s}|{:8s}|{:5s}|{:5s}'
_border = '--------------------------------------------'

# The commented out features are the ones that did not make the cut, while running models they actually
# ended up decreasing the accuracy which is why they have been excluded.

features = ([
    #             'pregnancies', 
                'glucose', 
                'bloodpressure', 
                'skinthickness', 
                'insulin',
                'bmi', 
                'diabetespedigreefunction', 
                'age', 
    #             'glucose_bins', 
    #              'bmi_bins',
    #              'age_bins',
                 'pregnancies_bins'
    #             'outcome'
                ])

# all results to be saved here
result_map = dict()
```

## SVM

```python
# define all the kernals here
kernals = {'linear': 'Linear SVM', 'rbf': 'Gaussian SVM', 'poly': 'Ploynomial SVM', 'sigmoid': 'Sigmoid SVM'}

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score


for kernal in kernals.keys():

    data = df.copy() #take a copy of existing read-in dataframe
    X = data[features]
    Y = data['outcome']

    # since SVM is a distance based calssifier, scaling values first
    scaler = StandardScaler()
    scaler.fit(X)
    X = scaler.transform(X)

    x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.5,random_state=3)
    
    #regularization parameter:
    reg = 1
    
    if kernal == 'poly': # account for degree=3 in case of poly SVM
        classifier = svm.SVC(kernel=kernal, degree=3, C=reg)
    else:
        classifier = svm.SVC(kernel=kernal, C=reg)
    y_pred = classifier.fit(x_train, y_train).predict(x_test)

    # results of Model
    sample = x_test.shape[0]
    result = calc_model_acc(y_test, y_pred, sample)
    print('**********************  {}  **********************'.format(kernals[kernal]))
    print(_border)
    print(_format.format('TP','FP','TN','FN','Accuracy','TPR','TNR'))
    print('{}\n{}\n{}'.format(_border,result,_border))
    
    # append the result in a dict 
    result_map.update({kernals[kernal]: result})
```

## Random Forest with optimal Depth and Estimator

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split


data = df.copy() #take a copy of existing read-in dataframe
X = data[features]
Y = data['outcome']
save_result = []
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.5,random_state=3)
# best combination from the above
estimator = 10; depth = 4

rf = RandomForestClassifier(n_estimators=estimator ,max_depth=depth, criterion ="entropy", random_state=3)
y_pred = rf.fit(x_train, y_train).predict(x_test)
error_rate = round((np.mean(y_pred != y_test)),4)
acc = 1 - error_rate
print('Estimator: {} Depth: {} Accuracy: {}'.format(estimator, depth, round(acc*100,2)))

# results of Model
sample = x_test.shape[0]
result = calc_model_acc(y_test, y_pred, sample)

print(_border)
print(_format.format('TP','FP','TN','FN','Accuracy','TPR','TNR'))
print('{}\n{}\n{}'.format(_border,result,_border))

result_map.update({'Random Forest(optimal config)': result})
```

### Picking the Best Model

Looking at the result from an accuracy stand-point Linear SVM and Logistic Regression produced the models with the highest accuracy. But in terms of just the True Negative Rate (TNR), Polynomial SVM prevailed even though it has lower overall accuracy in comparison to Linear SVM and Logistic Regression. Sigmoid SVM and Decision Tree yielded the models with the lowest accuracies. 



## Conclusion

There are two important facts about this dataset - first being it's small size and second being that it represents data from a particular tribe of people called Pima Indians. The chosen model should be used to predict if a person is diabetic or not from the same tribe since the distribution of data across all features is likely to be the same. The best model I would recommend is polynomial SVM (it has a degree=3), because this model has the highest TNR. Having the high TNR will ensure that patients are not misclassified as being diabetic based on their tests, this is important when dealing with remote tribes as arranging for tests can be quite an effort.
